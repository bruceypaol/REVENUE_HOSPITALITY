```python
###DATA TRANSFORMATION
```

```python
###New Column will be created and the formula will be run for each row
df_agg_bookings["occ_pct"] = df_agg_bookings['successful_bookings']/df_agg_bookings['capacity']
df_agg_bookings
```

```python
df_agg_bookings["occ_pct"] = df_agg_bookings["occ_pct"].apply(lambda x : round(x*100,2)) 
#orr by simple df_agg_bookings["occ_pct"] = round((df_agg_bookings['successful_bookings']/df_agg_bookings['capacity'])*100,2)
#in the previous steps
```

```python
#1.What is an average occupancy rate in each of the room categories?
df_agg_bookings.groupby("room_category")["occ_pct"].mean().round(2)
```

```python
# Since here we can only see the room category to get further info and clear picture about the room type we need to merge df_room
#and df_agg_bookings to get the same info in our previous query but with room details
df = pd.merge(df_agg_bookings, df_rooms, left_on = "room_category", right_on = "room_id") #Becoz common column have different name in both the table
df
```

```python
df.groupby("room_class")["occ_pct"].mean().round(2)
```

```python
#Now we can see what is the room type
#df have two similar columns room_id and room_category we can drop one
df.drop("room_id", axis=1, inplace= True)
df
```
